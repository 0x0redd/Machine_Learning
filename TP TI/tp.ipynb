```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lena.png')
rows, cols = img.shape[:2]
```

```python
M = np.float32([[1, 0, 50], [0, 1, 30]])
translated = cv2.warpAffine(img, M, (cols, rows))
```

```python
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Originale')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)), plt.title('Translat√©e')
plt.show()
```

```python
reduced = cv2.resize(img, (0,0), fx=0.5, fy=0.5)

zoomed = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
```

```python
center = (cols // 2, rows // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)), plt.title('Rotation 45¬∞')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lena.png')
img_reduced = cv2.resize(img, None, fx=0.5, fy=0.5)
img_enlarged = cv2.resize(img, None, fx=1.9, fy=1.9)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img_reduced, cv2.COLOR_BGR2RGB))
plt.title("R√©duction (√† 50%)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(img_enlarged, cv2.COLOR_BGR2RGB))
plt.title("Agrandissement (√† 190%)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
